```python
import pandas as pd
import matplotlib.pyplot as plt 
```

# Laptop Dataset analysis

```python
lap=pd.read_csv('Laptop-Price.csv')
```

```python
lap
```

```python
lap['Cpu Rate'] = lap['Cpu Rate'].str.replace('GHz', '').astype(float)
```

```python
lap['Ram'] = lap['Ram'].str.replace('GB', '').astype(int)
```

```python
lap
```

## Lets Check the price

```python
plt.figure(figsize=(10, 6))
plt.hist(lap['Price_euros'],color=(['green']),edgecolor='black')
print(len(lap['Price_euros']))
plt.title("Price distribution")
plt.xlabel("Price")
plt.ylabel("Laptops")
```

# Lets Check How many laptops there are with HDD and SSD

```python
## Counting the numbers of SSD
ssd_count=lap['SSD'].value_counts()

## Counting the numbers of HDD
hdd_count=lap['HDD'].value_counts()

## Combining the number of SSD and HDD into an single dataframe
combine=pd.DataFrame({
    'SSD':ssd_count,
    'HDD':hdd_count
}).fillna(0)

combine.plot(kind='bar',color=['blue','orange'])
plt.grid(True)
plt.title("Numrs of SSD and HDD of their respective size")
plt.xlabel("Size")
plt.ylabel("Count")
print(f"Total number of SSD is:{ssd_count}")
print(f"Total number of HDD is:{hdd_count}")
```
