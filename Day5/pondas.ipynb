## Reading the Csv file

```python
coffee= pd.read_csv('coffee.csv')
```

```python
coffee
```

## printing start to end

## .Loc is used to locate the rows this works as coffee.loc[rows,columns]

```python
coffee.loc[:]
```

## Printing the range of rows from 0 to end

```python
coffee.loc[0:]
```

## Printing a range of Rows

```python
coffee.loc[0:12]
```

## Printing Mulpiple lines

```python
coffee.loc[[0,1,2]]
```

### coffee.loc[[rows],[columns]]

```python
coffee.loc[[1,2,3,4],["Coffee Type","Units Sold"]]
```

If I want a specific value from the csv file such as tell me the number of sales on day 3 to get this type of single asnwer we can use .at

```python
coffee.at[3,"Units Sold"]
```

```python
### You can also do this 
coffee["Day"]
```

## If you want to see the csv file in sorted manner coffee.sort_values("column_name")

```python
coffee.sort_values("Units Sold")
```

### To sort in assending order we have to add 1 line extra is ascending=False coffee.sort_value("Units Sold", order=False)

```python
coffee.sort_values("Units Sold", ascending=False)
```
