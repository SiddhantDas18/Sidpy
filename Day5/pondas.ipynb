# Now we are working with another dataset


```python
olymp=pd.read_excel('olympics-data.xlsx')
```

```python
olymp
```

```python
olymp
```

## To locate  the specific value we can do that

```python
### We want to locate the specific values from the olympic list and lets do that where we can find the athelets who are more than 200cm

olymp.loc[olymp["height_cm"]>210]
```

```python
### Lets check if there is anyone from india

olymp.loc[olymp["born_country"]=="IND"]
```

```python
### Lets check if there is anyone from West bengal

olymp.loc[olymp["born_region"]=="West Bengal"]
```

## How to get some specific Columns attributes

```python
### Lets get the values who are from west bengal and their name nothing other than them

olymp.loc[olymp["born_region"]=="West Bengal",["name","born_region"]]
```

```python
olymp.sort_values("athlete_id")
```

```python
### We can also do sort hands to locate the cloums based on their born region

olymp[olymp['born_region']=="West Bengal"]
```

```python
## Lets say We only want their name sn their region

olymp[olymp['born_region']=="West Bengal"][['name','born_region']]
```

```python
### Lets say we want to filter the search we want two filters



olymp[(olymp['born_region']=="West Bengal") & (olymp['height_cm']>160)] [['name','born_region','height_cm']]
```

## If you want to sort them according to their name or what ever , only after you filter out the detils you need

```python
west=olymp[(olymp['born_region']=="West Bengal") & (olymp['height_cm']>160)] [['name','born_region','height_cm']]
west.sort_values("name")
```

## Using query inside pandas

```python

olymp.query('born_region=="West Bengal" & height_cm>160')
```
