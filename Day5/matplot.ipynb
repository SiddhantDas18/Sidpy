## Reading values from excel files and taking values to create  line chart

```python
sales=pd.read_excel('Book1.xlsx')
```

```python
sales
```

```python
a=sales['Day']
b=sales['Sales']
```

## To create a line chart

```python
plt.plot(a,b)
plt.grid(True)
plt.xlabel("Days")
plt.ylabel("Sales")
plt.title("Sales chart")
```

## To create a bar chart


```python
plt.bar(a,b)
```
